# Đệ quy

### Đệ quy là gì?

Đệ quy (recursion) là phương pháp một hàm tự gọi lại chính nó trong quá trình thực thi.

Một bài toán đệ quy cần có:
- Trường hợp cơ sở (base case) để dừng đệ quy.
- Công thức đệ quy (recursive case) để bài toán dần đơn giản hơn.

Ưu điểm:
- Code ngắn gọn, dễ hiểu với bài toán có tính chất chia nhỏ.

Nhược điểm:
- Nếu không có điều kiện dừng sẽ gây lỗi tràn stack.
- Có thể kém hiệu quả nếu không tối ưu (memoization).

Hãy sử dụng đệ quy để in 10 lần chữ hello

```python
def print_hello():
    print("hello")
    print_hello()

print_hello()
```

```python
def print_hello(i = 0):
    print("hello")
    if i < 9:
        print_hello(i+1)

print_hello()
```

```python
def print_hello(i = 0):
    print("hello", i)
    if i < 3:
        print_hello(i+1)
    print("hi", i)

print_hello()
```

```python
a = 10

def day_la_ham():
    # in biến a global (vì a không được khai báo trong hàm)
    print("in a bên trong hàm:", a)

day_la_ham()

print("in a bên ngoài hàm:", a)
```

```python
a = 10

def day_la_ham(a):
    # in biến a local (vì a tham số)
    print("in a bên trong hàm:", a)

day_la_ham(7)

print("in a bên ngoài hàm:", a)
```

```python
a = 10

def day_la_ham():
    # in biến a local (vì a được khai báo bên trong hàm)
    a = 7
    print("in a bên trong hàm:", a)

day_la_ham()

print("in a bên ngoài hàm:", a)
```

```python
a = 10

def day_la_ham():
    # in biến a local (vì a được khai báo bên trong hàm)
    # nhưng bị lỗi (vì a được khai báo sau khi in)
    print("in a bên trong hàm:", a)
    a = 7

day_la_ham()

print("in a bên ngoài hàm:", a)
```

```python
a = 10

def day_la_ham():
    # dùng keyword global nếu muốn thay đổi giá trị global
    # không phải tạo biến local mới
    global a
    a = 7
    print("in a bên trong hàm:", a)

day_la_ham()

print("in a bên ngoài hàm:", a)
```

Viết hàm `giai_thua` để tính giai thừa bằng cách sử dụng đệ quy

```python
def giai_thua(n):
    ket_qua = 1
    for i in range(1, n+1):
        ket_qua *= i
    return ket_qua

print(giai_thua(4))
```

```python
def giai_thua(n, i=1, ket_qua=1):
    if i <= n:
        ket_qua_tang_tiep_theo = giai_thua(n, i+1, ket_qua*i)
        return ket_qua_tang_tiep_theo
    return ket_qua

print(giai_thua(4))
```

```python
def giai_thua(n):
    if n == 1:
        return 1
    return n * giai_thua(n-1)

print(giai_thua(4))
```

### Bài 2

Viết hàm `fibonacci` có một tham số đầu vào là số nguyên n, hãy in n số fibonacci đầu tiên (bắt đầu từ 1).

```python
# 1 1 2 3 5 8 13

# S = 1 + 2 + ... + n

# fibonacci thứ n = fibonacci thứ (n - 1) + fibonacci thứ (n - 2)

def fibonacci(n):
    if n <= 2:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(5))
print(fibonacci(6))
print(fibonacci(7))
```

```python
def fibonacci(n, a = 1, b = 0, i = 1):
    if i < n:
        return fibonacci(n, a + b, a, i+1)
    return a

print(fibonacci(5))
```
