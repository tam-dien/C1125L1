### Bài 1

Viết hàm với đầu vào là một mảng, hãy tính và trả về giá trị trung bình của mảng

```python
def gia_tri_tb(mang):
    if len(mang) == 0:
        return 0
        
    tong = 0
    for i in mang:
        tong += i
    return tong/len(mang)

mang = [2,3,4,5,6,7,-8,9]
print(gia_tri_tb(mang))
```

### Bài 2

Viết hàm với đầu vào là một mảng, hãy tìm và trả về giá trị âm lớn nhất trong mảng

```python
def so_am_lon_nhat(lst):
    max_am = None
    for x in lst:
        if x < 0:
            if max_am is None or x > max_am:
                max_am = x
    return max_am

print(so_am_lon_nhat([32, 3, 4, 99, 8, 22]))
print(so_am_lon_nhat([32, 3, 4, -7, 8, 22]))
print(so_am_lon_nhat([32, 3, 4, -7, 8, -1]))
```

### Bài 3

Viết hàm với giá trị đầu vào là một mảng, hãy sắp xếp và trả về mảng có độ lớn tăng dần

```python
# Vân

def sap_xep_do_tang(list_a):
    list_a = list_a.copy()
    list_b = []
    while len(list_a) != 0:
        min = list_a[0]
        for i in list_a:
            if i < min:
                min = i
        list_b.append(min)
        list_a.remove(min)
    return list_b

list_so = [15,7,15,84,-3]
ket_qua = sap_xep_do_tang(list_so)

print("list_so:", list_so)
print("Kết quả:", ket_qua)
```

```python
def sap_xep_do_tang(list_a):
    list_a = list_a.copy()
    for i in range(len(list_a)-1):
        for j in range(i+1, len(list_a)):
            if list_a[i] > list_a[j]:
                list_a[i], list_a[j] = list_a[j], list_a[i]
    return list_a

list_so = [15,7,15,84,-3]
ket_qua = sap_xep_do_tang(list_so)

print("list_so:", list_so)
print("Kết quả:", ket_qua)
```

### Bài 4

Viết hàm với giá trị đầu vào là một mảng, tìm x sao cho đoạn [-x,x] chứa tất cả các giá trị có trong mảng

```python
def tim_x(list_a):
    x = abs(list_a[0])
    for i in list_a:
        if abs(i) > x:
            x = abs(i)
    return x

print(tim_x([12, 10, 8, -7]))
print(tim_x([12, 10, 7, -20]))
```

### Bài 5

Viết hàm với giá trị đầu vào là một mảng, Hãy tìm một vị trí trong mảng thỏa hai điền kiện: có hai giá trị lân cận và giá trị tại vị trí đó bằng tích hai giá trị lân cận. Nếu mảng không tồn tại giá trị như vậy thì hàm trả về giá trị - 1

```python
def tim_so(list_a):
    for i in range(1,len(list_a)-1):
        if list_a[i] == list_a[i-1]*list_a[i+1]:
            return i
    return -1
list_a = [7,5,-2,6,24,4]
print(tim_so(list_a))
```

### Bài 6

Viết hàm có giá trị đầu vào là một mảng. Hãy tìm giá trị đầu tiên trong mảng một chiều các số nguyên có chữ số đầu tiên là chữ số lẻ. Nếu trong mảng không tồn tại giá trị như vậy thì hàm sẽ trả về giá trị 0.

```python
def tim_gia_tri(mang): 
    for i in mang:
        if i < 0:
            so = -i
        else:
            so = i
        chu_so_dau = int(str(so)[0])
        if chu_so_dau % 2 != 0:
            return i
    return 0

mang = [8, 4, 45, -7, 28, -83, 22]
print(tim_gia_tri(mang))
mang = [8, 4, 15, -7, 28, -83, 22]
print(tim_gia_tri(mang))
```

### Bài 7

Viết hàm có giá trị đầu vào là một mảng. Hãy tìm giá trị đầu tiên trong mảng có dạng 2^k. Nếu mảng không tồn tại giá trị dạng 2^k thì hàm sẽ trả về giá trị 0.

```python
def kiem_tra_2_mu_k(a):
    if a < 1:
        a = 1/a
    while a >= 2:
        a /= 2

    return a == 1

def tim_gia_tri(list_a):
    for i in list_a:
        if kiem_tra_2_mu_k(i):
            return i
    return 0

list_a = [3,5,0.5,4,6]
print(tim_gia_tri(list_a))
```

### Bài 8

Viết hàm có giá trị đầu vào là một mảng. Hãy tìm giá trị thỏa điều kiện toàn chữ số lẻ và là giá trị lớn nhất thỏa điều kiện ấy trong mảng (nếu mảng không có giá trị thỏa điều kiện trên thì hàm trả về giá trị 0)

```python
def chu_so_le(a):
    for ky_tu in str(a):
        if int(ky_tu) % 2 == 0:
            return False
    return True
    
def func(list_a):
    ket_qua = 0
    for phan_tu in list_a:
        if chu_so_le(phan_tu) and phan_tu > ket_qua:
            ket_qua = phan_tu
    return ket_qua

arr = [2,21,33,3,35,63,71,87]
print(func(arr))
```
