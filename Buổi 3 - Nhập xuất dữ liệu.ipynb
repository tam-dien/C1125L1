### Xuất dữ liệu

Lệnh `print()` là lệnh giúp python xuất dữ liệu ra màn hình

Cú pháp:

`print(*object, sep=' ', end='\n')`

Trong đó:

- objects – là các giá trị sẽ được xuất ra màn hình
- sep – phân cách giữa các giá trị. Giá trị mặc định của sep là một khoảng trắng.
- end – là kí tự được in ra sau khi các giá trị của objects đã được in. Mặc định của end là ‘\n’ – xuống dòng mới.
##### objects có thể là nhiều giá trị, mỗi giá trị cách nhau một dấu phẩy

Ví dụ:

`print("Hello","mọi người")`

Ở đây, objects là "HEllo" và "mọi người"

##### Đọc thêm tài liệu tại đây: https://www.w3schools.com/python/ref_func_print.asp

### Nhập dữ liệu

Lệnh `input()` là lệnh giúp python nhận dữ liệu nhập từ bàn phím

Cú pháp:

`input(prompt)`

Trong đó:

- prompt là nội dung in ra màn hình

Giá trị trả về của lệnh `input()` mang kiểu dữ liệu là String (chuỗi)

Thực hiện doạn lệnh yêu cầu nhập vào tên người dùng, sau đó in ra dòng chữ "Xin chào [tên người dùng]":

```python
ket_qua = input("Nhập vào a:")
phep_cong = ket_qua + 10

print(phep_cong)
```

```python
ket_qua = input("Nhập vào a:")
ket_qua = int(ket_qua)
phep_cong = ket_qua + 10

print(phep_cong)
```

```python
ket_qua = int(input("Nhập vào a:"))
phep_cong = ket_qua + 10

print(phep_cong)
```

```python
ket_qua = int(float(input("Nhập vào a:")))
phep_cong = ket_qua + 10

print(phep_cong)
```

```python
ket_qua = float(input("Nhập vào a:"))
phep_cong = ket_qua + 10

print(phep_cong)
```

Thực hiện một đoạn lệnh yêu cầu nhập vào bán kính một đường tròn, sau đó in ra diện tích và chu vi hình tròn đó

```python
r = float(input("r:"))

Chu_vi = 2*r*3.14
Dien_tich = (r**2)*3.14

print(Chu_vi)
print(Dien_tich)
```

# Bài tập

Nhập vào số a.

Hãy kiểm tra xem a có phải là một số nguyên hay không.

Nếu a là số nguyên, in True
Ngược lại, in False

```python
a = float(input("Nhập vào a:"))

### ví dụ: người dùng nhập 7.1 ~~> False
###        người dùng nhập 7 ~~> True
ket_qua = a == int(a)
print(a)
print(ket_qua)
```

Nhập vào số a.

Hãy kiểm tra xem a có phải là một số chính phương hay không.

Số chính phương là một số tự nhiên mà căn bậc hai của nó là một số tự nhiên.

Nếu a là số chính phương, in True
Ngược lại, in False

```python
a = float(input("Nhập vào a:"))
can_a = a**0.5

kq = can_a == int(can_a)
print(a)
print(kq)
```

Nhập vào các số a, b và x.

Hãy in xem, x có thuộc trong khoảng từ a đến b (bao gồm cả a và b).

Nếu có, in True.
Ngược lại, in False.

```python
a = float(input("Nhập vào a:"))
b = float(input("Nhập vào b:"))
x = float(input("Nhập vào x:"))

ket_qua = x >= a and x <= b
print("a:", a)
print("b:", b)
print("x:", x)
print(ket_qua)
```

```python
a = float(input("Nhập vào a:"))
b = float(input("Nhập vào b:"))
x = float(input("Nhập vào x:"))

ket_qua = a <= x <= b
print("a:", a)
print("b:", b)
print("x:", x)
print(ket_qua)
```

Nhập vào số a.

Nếu a lớn hơn 10, in "Số lớn hơn 10", ngược lại, in "Số không lớn hơn 10".

<span style="color:#ff4d4f">* Không sử dụng các lệnh rẽ nhánh như if, for, while,....</span>

```python
a = float(input("Nhập vào a:"))

ket_qua = (a > 10 and "Số lớn hơn 10") or "Số không lớn hơn 10"

print(a)
print(ket_qua)
```

Nhập vào số a.

Hãy kiểm tra xem a có chia hết cho 3 hay không.

Nếu a chia hết cho 3, hãy in "Số chia hết cho 3" ra màn hình.

Nếu a không chia hết cho 3, hãy in "Số không chia hết cho 3" ra màn hình.

<span style="color:#ff4d4f">* Không sử dụng các lệnh rẽ nhánh như if, for, while,....</span>

```python
a = float(input("Nhập vào a:"))

ket_qua = (a % 3 == 0 and "Số chia hết cho 3") or "Số không chia hết cho 3"

print(a)
print(ket_qua)
```

nhập vào số a và số b, hãy giải phương trình ax + b = 0.

In x lên màn hình theo mẫu như ví dụ.

Trong trường hợp a bằng 0, in "a bằng 0".

<span style="color:#ff4d4f">* Không sử dụng các lệnh rẽ nhánh như if, for, while,....</span>

```python
a = float(input("Nhập vào a:"))
b = float(input("Nhập vào b:"))

ket_qua = (a == 0 and "a bằng 0") or -b/a

print(a)
print(b)
print(ket_qua)
```
