### Bài 1

Cho phương trình bậc 2: ax^2 + bx + c = 0

Nhập a, b, c

Giải và biện luận phương trình trên

```python
a = float(input("Nhập vào số a:"))
b = float(input("Nhập vào số b:"))
c = float(input("Nhập vào số c:"))

print("Phương trình: ",a,"x^2 + ",b,"x + ",c," = 0", sep="")

if a == 0:
    if b == 0:
        if c == 0:
            print("Phương trình có vô số nghiệm")
        else:
            print("Phương trình vô nghiệm")
    else:
        x = -c/b
        print("Phương trình có 1 nghiệm x =", x)
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print("Phương trình vô nghiệm")
    elif delta == 0:
        x = -b/2*a
        print("Phương trình có nghiệm kép x =", x)
    else:
        can_delta = delta**0.5
        x1 = (-b + can_delta)/2*a
        x2 = (-b - can_delta)/2*a
        print("Nghiệm thứ nhất là x1 =", x1)
        print("Nghiệm thứ 2 là x2 =", x2)
```

Tại một cửa hàng nọ có menu như sau:

1. Gà rán giá 35k
2. Hamburger giá 50k
3. Pesi giá 10k

Nhập số lượng món hàng cần mua

Tính và in ra tổng giá tiền (hiển thị có dấu chấm mỗi 3 chữ số để người dùng để quan sát)

Nếu giá trị đơn hàng trên 2 triệu, giảm 10% cho khách, trên 4 triệu, giảm 15%

Tính và in ra tổng giá tiền được giảm (hiển thị có dấu chấm mỗi 3 chữ số để người dùng để quan sát)

Tính và in ra tổng giá tiền khách cần trả (hiển thị có dấu chấm mỗi 3 chữ số để người dùng để quan sát)

In ra dòng chữ "Cảm ơn quý khách đã mua hàng"

```python
ga = int(input('Số lượng gà:'))
ham = int(input('Số lượng Hambuger:'))
pep = int(input('Số lượng Pepsi:'))

tong = ga*35000 + ham*50000 + pep*10000

print('Tổng tiền','{:,}'.format(tong).replace(',','.'))
if tong > 4_000_000:
    giam = tong*0.15
    print('Bạn được giảm 15%')
    print('Số tiền được giảm','{:,}'.format(int(giam)).replace(',','.'))
elif tong > 2_000_000:
    giam = tong*0.1
    print('Bạn được giảm 10%')
    print('Số tiền được giảm','{:,}'.format(int(giam)).replace(',','.'))
else:
    giam = 0

tra = tong - giam
print('Số tiền phải trả:','{:,}'.format(int(tra)).replace(',','.'))
```

### Bài 6

Nhập vào chiều cao (cm) và cân nặng (kg), tính số BMI và xét rồi xin kết quả theo dữ liệu sau:

- BMI < 16: Gầy cấp độ III
- 16 <= BMI < 17:  Gầy cấp độ II
- 17<= BMI < 18.5: Gầy cấp độ I
- 18.5 <= BMI < 25: Bình thường
- 25 <= BMI < 30: Thừa cân
- 30 <= BMI < 35 : Béo phì cấp độ I
- 35 <= BMI < 40: Béo phì cấp độ II
- BMI > 40: Béo phì cấp độ III

```python
w = float(input("Nhập vào cân nặng (kg): "))
h = float(input("Nhập vào chiều cao (cm): "))
BMI = w / ((h/100) ** 2)

if BMI < 16:
    print("gầy cấp độ III")
elif BMI < 17: 
    print("Gầy cấp độ II")
elif BMI < 18.5:
    print("Gầy cấp độ I")
elif BMI < 25:
    print("Bình thường")
elif BMI < 30:
    print("Thừa cân")
elif BMI < 35:
    print("Béo phì cấp độ I")
elif BMI < 40:
    print("Béo phì cấp độ II")
else:
    print("Béo phì cấp độ III")
```

```python
ngay = int(input("Nhập ngày: "))
thang = int(input("Nhập tháng: "))
nam = int(input("Nhập năm: "))

so_ngay = 0
# so_ngay = ngay + so_thang_31_ngay*31 + so_thang_co_30_ngay*30

# ví dụ:
# nhập ngày 14/7/1992
# so_ngay = 14 + 3*31 + 3*30 - 1

# ví dụ:
# nhập ngày 5/1/1992
# so_ngay = 5

# thang | so_thang_31_ngay | so_thang_30_ngay
# 1 | 0 | 0
# 2 | 1 | 0
# 3 | 1 | 1
# 4 | 2 | 1
# 5 | 2 | 2
# 6 | 3 | 2
# 7 | 3 | 3
# 8 | 4 | 3
# 9 | 5 | 3
# 10 | 5 | 4
# 11 | 6 | 4
# 12 | 6 | 5

if thang <= 7:
    so_thang_31_ngay = thang//2
    so_thang_30_ngay = (thang-1)//2
else:
    so_thang_31_ngay = (thang+1)//2
    so_thang_30_ngay = (thang-2)//2

so_ngay = ngay + so_thang_31_ngay*31 + so_thang_30_ngay*30
if thang > 2:
    if nam % 4 == 0:
        so_ngay -= 1
    else:
        so_ngay -= 2

print(ngay,"/",thang,"/",nam,sep="")
print(so_ngay)
```
