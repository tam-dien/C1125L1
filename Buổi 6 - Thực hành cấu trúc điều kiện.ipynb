### Bài 1

Cho phương trình bậc 2: ax^2 + bx + c = 0

Nhập a, b, c

Giải và biện luận phương trình trên

```python
a = float(input("Nhập vào số a:"))
b = float(input("Nhập vào số b:"))
c = float(input("Nhập vào số c:"))

print("Phương trình: ",a,"x^2 + ",b,"x + ",c," = 0", sep="")

if a == 0:
    if b == 0:
        if c == 0:
            print("Phương trình có vô số nghiệm")
        else:
            print("Phương trình vô nghiệm")
    else:
        x = -c/b
        print("Phương trình có 1 nghiệm x =", x)
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print("Phương trình vô nghiệm")
    elif delta == 0:
        x = -b/2*a
        print("Phương trình có nghiệm kép x =", x)
    else:
        can_delta = delta**0.5
        x1 = (-b + can_delta)/2*a
        x2 = (-b - can_delta)/2*a
        print("Nghiệm thứ nhất là x1 =", x1)
        print("Nghiệm thứ 2 là x2 =", x2)
```
