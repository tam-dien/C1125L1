### Hàm


Gọi hàm:
`<tên hàm>(<giá trị đầu vào>)`

Ví dụ:

Hãy viết 1 hàm số để tính tổng 2 số đầu vào và in nó ra màn hình

````python
def in_menu(in_pizza): # in_pizza gọi là parameter
    print("------------------")
    print("Menu:")
    print("Gà rán: 40k")
    print("Pesi: 10k")
    if in_pizza:
        print("Pizza: 50k")

in_menu(True) # True gọi là argument
in_menu(False)
in_menu(True)
in_menu(False)
in_menu(True)
````

````python
# viết hàm in_menu, và thực hiện in_menu 6 lần
# ở lần 1 và 3 in đầy đủ menu
# ở lần 2 và 4 không in pizza
# các lần còn lại, không in pesi

def in_menu(in_pizza, in_pesi): # in_pizza gọi là parameter
    print("------------------")
    print("Menu:")
    print("Gà rán: 40k")
    if in_pesi:
        print("Pesi: 10k")
    if in_pizza:
        print("Pizza: 50k")

in_menu(True, True)
in_menu(False, True)
in_menu(True, True)
in_menu(False, True)
in_menu(True, False)
in_menu(True, False)
````

````python
# viết hàm tim_max, truyền vào được tối đa 5 tham số
# khi gọi hàm chúng ta có thể truyền vào từ 2 đến 5 giá trị
# tìm giá trị max

# tim_max(5, 8, 10, 15, 5) ~~> 15
# tim_max(5, 8, 15, 5) ~~> 15
# tim_max(5, 8, 5) ~~> 8
# tim_max(5, 10) ~~> 10

# tim_max(-10, -5, -2, -1, -15) ~~> -1

def tim_max(a, b, c=None, d=None, e=None):
    so_lon_nhat = a
    if so_lon_nhat < b:
        so_lon_nhat = b
    if c != None and so_lon_nhat < c:
        so_lon_nhat = c
    if d != None and so_lon_nhat < d:
        so_lon_nhat = d
    if e != None and so_lon_nhat < e:
        so_lon_nhat = e
    print(so_lon_nhat)

tim_max(5, 8, 10, 15, 5)
tim_max(5, 8, 15, 5)
tim_max(5, 8, 5)
tim_max(5, 10)
````

````python
def tim_max(a, b, c=None, d=None, e=None):
    so_lon_nhat = a
    if so_lon_nhat < b:
        so_lon_nhat = b
````
````python
    if c == None:
        return so_lon_nhat
    
    if so_lon_nhat < c:
        so_lon_nhat = c
    if d == None:
        return so_lon_nhat
    
    if so_lon_nhat < d:
        so_lon_nhat = d
    if e == None:
        return so_lon_nhat
    
    if so_lon_nhat < e:
        so_lon_nhat = e
    return so_lon_nhat

so_1 = tim_max(5, 8, 10, 15, 5)
so_2 = tim_max(5, 8, 15, 5)
so_3 = tim_max(5, 8, 5)
so_4 = tim_max(5, 10)

print(so_1)
print(so_2)
print(so_3)
print(so_4)
````

### Bài tập

Viết hàm kiểm tra số nguyên tố

````python
# viết hàm so_nguyen_to có 1 tham số là a
# nếu a là số nguyên tố thì trả về True, ngược lại trả về False

def so_nguyen_to(a):
    if a < 2:
        return False
    for i in range(2, a):
        if a % i == 0:
            return False
    return True

ket_qua_0 = so_nguyen_to(2)
print(ket_qua_0) # True
ket_qua_1 = so_nguyen_to(7)
print(ket_qua_1) # True
ket_qua_2 = so_nguyen_to(9)
print(ket_qua_2) # False
ket_qua_3 = so_nguyen_to(23)
print(ket_qua_3) # True
````

Viết hàm đếm số lượng từ của một chuỗi đưa vào

````python
# chuoi đưa vào có dấu câu, giữa các từ có thể có nhiều hơn 1 khoảng trắng

# dem_tu("xin chào mọi người") ~~> 4 từ
# dem_tu("xin   chào   mọi  người") ~~> 4 từ
# dem_tu("  xin   chào   mọi  người  ") ~~> 4 từ

def dem_tu(chuoi: str):
    chuoi += " "
    dem = 0
    tach_chuoi = ""
    for ky_tu in chuoi:
        if ky_tu != " ":
            tach_chuoi += ky_tu
        else: 
            if tach_chuoi != "":
                dem += 1
                tach_chuoi = ""
    return dem

dem_tu("xin     ba ly   nhe ")

print(dem_tu("xin chào mọi người"))
print(dem_tu("xin   chào   mọi  người"))
print(dem_tu("  xin   chào   mọi  người  "))
````
