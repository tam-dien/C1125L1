# Xử lý file

### Mở file

Lệnh mở file:

`f = open(file_path, mode)`

Sau khi xử lý xong, luôn luôn phải đóng file lại để tránh những lỗi không mong muốn.

`f.close()`

Sử dụng with để tự động đóng file:

```
with open(file_path, mode) as f:
    ## xử lý file
```

##### Lưu ý, để xử lý được tiếng việt, ta cần thêm tham số encoding="utf-8"

### Đọc file

Sử dụng mode `r` để đọc file.

Các lệnh đọc file:

#### `f.read()`

Lệnh này trả về giá trị kiểu dữ liệu string.

Lệnh này trả về toàn bộ nội dung file.

````python
f = open("test.txt", encoding="utf-8")
noi_dung = f.read()
f.close()

print(noi_dung)
````

````python
f = open("xinchao.docx", encoding="utf-8")
noi_dung = f.read()
f.close()

print(noi_dung)
````

#### `f.readline()`

Lệnh này trả về giá trị kiểu dữ liệu string.

Lệnh này trả về một dòng dữ liệu trong file.

````python
f = open("test.txt", encoding="utf-8")
noi_dung = f.readline()
noi_dung_2 = f.readline()
noi_dung_3 = f.readline()
f.close()

print(noi_dung)
print(noi_dung_2)
print(noi_dung_3)
````

#### `f.readlines()`

Lệnh này trả về giá trị kiểu dữ liệu list các string.

Lệnh này trả về một list các dòng dữ liệu trong file.

````python
f = open("test.txt", encoding="utf-8")
noi_dung = f.readlines()
f.close()

print(noi_dung)
````

````python
f = open("sinh_vien.csv", encoding="utf-8")
noi_dung = f.readlines()
f.close()

print(noi_dung)
danh_sach_sinh_vien = [
    {"Mã SV": "0001", "Tên SV": "Phạm Ngu..."},
    {"Mã SV": "0002", "Tên SV": "Cao Mỹ Vân..."},
]

### viết hàm với đầu vào là 1 đường dẫn file, đọc file và
### trả về dữ liệu theo dạng JSON (list of dictionary)
````

````python
def doc_file(duong_dan_file):
    f = open(duong_dan_file, encoding="utf-8")
    tieu_de = f.readline()
    noi_dung = f.readlines()
    f.close()

    ket_qua = []

    tieu_de = tieu_de.strip().split(",")
    for dong in noi_dung:
        dong = dong.strip().split(",")
        dong_moi = {}
        for i in range(len(tieu_de)):
            ten_cot = tieu_de[i]
            dong_moi[ten_cot] = dong[i]
        ket_qua.append(dong_moi)

    return ket_qua

du_lieu = doc_file("sinh_vien_C1125L1.csv")
# lọc ra những học viên có thời gian nhập học trong tháng 09
list_loc = []
for sinh_vien in du_lieu:
    if sinh_vien["Nhập học"][3:5] == "09":
        list_loc.append(sinh_vien)

print(list_loc)
````

### Ghi file

Sử dụng mode `w` để viết file lại từ đầu, hoặc dùng mode `a` để thêm dữ liệu mới vào file.

Lệnh ghi file:

#### `f.write(content)`

````python
f = open("sinh_vien_C1125L1.csv", "w", encoding="utf-8")
f.write("abc")
f.close()
````

````python
f = open("sinh_vien_C1125L1.csv", "a", encoding="utf-8")
f.write("abc")
f.close()
````

````python
f = open("sinh_vien_C1125L1.csv", "a", encoding="utf-8")
f.write("CGO008504,Nguyễn Văn A,Đang học,01-01-2026\n")
f.close()
````

````python
def ghi_file(duong_dan_file, du_lieu):
    ## du_lieu: list of dictionary
    ## cần chuyển về dạng csv để ghi file
    if len(du_lieu) == 0:
        return
    tieu_de = ""
    for key in du_lieu[0]:
        tieu_de += key + ","
    tieu_de = tieu_de[:-1]
    f = open(duong_dan_file, "w", encoding="utf-8")
    f.write(tieu_de)
    f.close()

du_lieu = [
    {"Mã học viên": "CGO008266", "Họ tên": "Cao Anh Phong", "Trạng thái": "Đang học", "Nhập học": "01-01-2026"},
    {"Mã học viên": "CGO008036", "Họ tên": "Đặng Quang Dũng", "Trạng thái": "Đang học", "Nhập học": "01-01-2026"},
]
ghi_file("sinh_vien.csv", du_lieu)
````
