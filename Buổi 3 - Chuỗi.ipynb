### Chuỗi dữ liệu

Khai báo chuỗi:
    
`chuoi = "hello, mọi người!!"`

Trong python, chuỗi có thể được đặt trong hai dấu nháy đơn, hoặc hai dấu nháy kép

### Truy xuất ký tự trong chuỗi

Cú pháp:

`chuỗi[chỉ số]`

Chỉ số bắt đầu là 0

Chỉ số cuối cùng của chuỗi là -1

Ví dụ:

```python
chuoi = "hello mọi người"
print(chuoi)
```

```python
chuoi = 'hello mọi người'
print(chuoi)
```

```python
chuoi = 'hello mọi người'
ky_tu = chuoi[14]
print(ky_tu)
```

```python
chuoi = 'hello mọi người'
ky_tu = chuoi[-1]
print(ky_tu)
```

```python
chuoi = 'hello mọi người'
do_dai = len(chuoi)

ky_tu = chuoi[-15]
print(ky_tu)
```

```python
### Nhập vào 1 chuỗi bất kỳ
### kiểm tra xem ký tự thứ 5 từ cuối chuỗi có phải là ký tự n hay không
### nếu có thì in True, nếu không thì in False

chuoi = input("Nhập chuỗi:")
ket_qua = len(chuoi) > 4 and "n" == chuoi[-5]

## nếu số lượng ký tự ít hơn 5, chương trình sẽ lỗi
## sửa lại: nếu số lượng ký tự ít hơn 5 ~> in False

print(chuoi)
print(ket_qua)
```

### Cắt chuỗi

Cú pháp:

`chuỗi[a: b]`

Trong đó
- a là vị trí đầu
- b là vị trí cuối +1


```python
chuoi = 'hello mọi người'
chuoi_2 = chuoi[0:7]

print(chuoi_2)
```

```python
chuoi = 'hello mọi người'
chuoi_2 = chuoi[2:7]

print(chuoi_2)
```

```python
chuoi = 'hello mọi người'
chuoi_2 = chuoi[:7]

print(chuoi_2)
```

```python
chuoi = 'hello mọi người'
chuoi_2 = chuoi[2:]

print(chuoi_2)
```

```python
chuoi = 'hello mọi người'
chuoi_2 = chuoi

print(chuoi_2)
```

```python
### nhập vào chuỗi, hãy kiểm tra xem 5 ký tự cuối cùng của chuỗi có phải là abcde hay không
### nếu có thì in True, nếu không thì in False

chuoi = input("Nhập chuỗi:")

ket_qua = chuoi[-5:] == "abcde"
print(chuoi)
print(ket_qua)
```

### Toán tử với chuỗi

#### Toán tử +

Toán tử + giúp gắn liền 2 chuỗi:

Ví dụ:

```python
chuoi_a = "xin chào"
chuoi_b = "các bạn"

chuoi_c = chuoi_a + " " + chuoi_b
print(chuoi_c)
```

#### Toán tử *

Toán tử * giúp lặp lại đoạn chuỗi

Ví dụ:

```python
chuoi_a = "xin chào"
so_nguyen = 7

chuoi_b = chuoi_a * so_nguyen
print(chuoi_b)
```

```python
chuoi_a = "xin chào"
so_nguyen = 7.5

chuoi_b = chuoi_a * so_nguyen
print(chuoi_b)
```

### Một số lệnh xử lý String

#### len – trả về độ dài của chuỗi

Cú pháp:

`len(chuỗi)`

Kết quả trả về có kiểu dữ liệu là int (kiểu số nguyên)

Ví dụ:

```python
chuoi_a = "xin chào"
do_dai = len(chuoi_a)

print(do_dai)
```

```python
chuoi_a = "xin \nchào"
do_dai = len(chuoi_a)

print(do_dai)
```

####  count – hàm trả về số lần chuỗi con có mặt trong chuỗi

Cú pháp:

`chuỗi.count(value)`

Trong đó:

- valve là giá trị chuỗi cần tìm

Kết quả trả về có kiểu dữ liệu là int (kiểu số nguyên)

Ví dụ:

```python
chuoi_a = "xin chào các bạn, xin chào mọi người"
chuoi_b = "n"

so_lan = chuoi_a.count(chuoi_b)
print(so_lan)
```

```python
chuoi_a = "xin chào các bạn, xin chào mọi người"
chuoi_b = "xin chào"

so_lan = chuoi_a.count(chuoi_b)
print(so_lan)
```

```python
chuoi_a = "xin chào các bạn, xin chào mọi người"
chuoi_b = "ạ"

so_lan = chuoi_a.count(chuoi_b)
print(so_lan)
```

```python
chuoi_a = "xin chào các bạn, xin chào mọi Người"
chuoi_b = "n"

so_lan = chuoi_a.count(chuoi_b)
print(so_lan)
```

##### *1KB = 1024 Byte
##### *1MB = 1024 KByte
##### *1GB = 1024 MByte
##### *1TB = 1024 GByte
