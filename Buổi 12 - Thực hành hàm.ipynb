Viết hàm có dữ liệu đầu vào là 2 số a và b, số nào lớn hơn thì in bảng cửu chương của số đó

```python
def in_bang_cuu_chuong(a,b):
    if a > b:
        print("Bảng cửu chương số:", a)
        for i in range(1,11):
            print(a, "x", i, "=", a*i)
    else:
        print("Bảng cửu chương số:", b)
        for i in range(1,11):
            print(b, "x", i, "=", b*i)
    
in_bang_cuu_chuong(5,8)
```

### Bài 2

Viết hàm `xoa_chuoi` có dữ liệu đầu vào là 2 chuỗi a và b, hãy tìm trong chuỗi a có chuỗi b hay không, nếu có hãy xóa toán bộ chuỗi b trong a.

Sau đó trả về chuỗi a mới

Ví dụ:

chuoi1 = "xin chào các bạn, các bạn có khỏe không?"

chuoi2 = "các bạn"

chuoi3 = xoa_chuoi(chuoi1,chuoi2)

print(chuoi3)

Kết quả in ra là:
xin chào ,  có khỏe không

```python
def xoa_chuoi(a, b):
    ket_qua = ""
    i = 0

    while i < len(a):
        if a[i: i+len(b)] == b:
            i += len(b)
        else:
            ket_qua += a[i]
            i += 1

    return ket_qua

chuoi1 = "Xin chào các bạn, các bạn có khỏe không?"
chuoi2 = "các bạn"
chuoi3 = xoa_chuoi(chuoi1,chuoi2)
print(chuoi3)
```

```python
def xoa_chuoi(a, b):
    return a.replace(b, "")
    
chuoi1 = "Xin chào các bạn, các bạn có khỏe không?"
chuoi2 = "các bạn"
chuoi3 = xoa_chuoi(chuoi1,chuoi2)
print(chuoi3)
```

### Bài 3

Viết hàm `kiemtramatkhau` với dữ liệu đầu vào là 1 chuỗi có công dụng như sau:

Kiểm tra xem chuỗi đầu vào có phải là mật khẩu mạnh hay không, nếu có trả về True, nếu không trả về False

Mật khẩu mạnh là mật khẩu phải có ít nhất 1 chữ hoa, 1 chữ thường, 1 số và 1 ký tự đặc biệt

```python
def kiem_tra_mat_khau(a):
    check_hoa = False
    check_thuong = False
    check_so = False
    check_ktdb = False
    for i in a:
        if i.isupper():
            check_hoa = True
        elif i.islower():
            check_thuong = True
        elif i.isdigit():
            check_so = True
        else:
            check_ktdb = True

        if check_hoa and check_thuong and check_so and check_ktdb:
            return True
    return False

print(kiem_tra_mat_khau("Ask@dj23"))
print(kiem_tra_mat_khau("ask@dj23"))
print(kiem_tra_mat_khau("ASK@DJ23"))
print(kiem_tra_mat_khau("Askadj23"))
print(kiem_tra_mat_khau("Ask@djab"))
```

### Bài 4

Viết hàm có 5 dữ liệu đầu vào là 5 số nguyên, tìm giá trị lớn nhì trong 5 số đó

```python
def gia_tri_lon_nhi(a,b,c,d,e):
    so_lon_nhat = a
    so_lon_nhi = b
    
    if so_lon_nhi > so_lon_nhat:
        so_lon_nhat, so_lon_nhi = so_lon_nhi, so_lon_nhat

    if c > so_lon_nhat:
        so_lon_nhi = so_lon_nhat
        so_lon_nhat = c     
    if d > so_lon_nhat :
        so_lon_nhi = so_lon_nhat
        so_lon_nhat = d
    if e > so_lon_nhat:
        so_lon_nhi = so_lon_nhat
        so_lon_nhat = e 
    return so_lon_nhi

x = gia_tri_lon_nhi(2,7,6,9,15)
print("Số lớn nhì:", x)

x = gia_tri_lon_nhi(14,18,6,10,15)
print("Số lớn nhì:", x)
```

### Bài 5

Viết hàm kiểm tra số nguyên tố

Viết hàm kiểm tra số hoàn hảo

Viết hàm kiểm tra số chính phương

Viết hàm kiểm tra số Armstrong

Nhập vào một chuỗi, hãy lọc và tìm ra những số ở trong chuỗi (2 chữ số liền nhau là một chuỗi)

Ví dụ chuỗi: "AB25SDc5DSjjh52"

Trong chuỗi này ta có 3 số: 25, 5 và 52

Sau khi lọc số, dùng các hàm trên để kiểm tra và thực hiện như sau:

Cho biến S

- Nếu số đó là số nguyên tố, ta thực hiện S += số đó
- Nếu số đó là số hoàn hảo, ta thực hiện S -= số đó
- Nếu số đó là số chính phương, ta thực hiện S /= số đó
- Nếu đó là số Armstrong ta cho S = 0

Hãy tính và in S

```python
def so_nguyen_to(a):
    if a < 2:
        return False
    for i in range(2, a):
        if a % i == 0:
            return False
    return True
    
def so_hoan_hao(b):
    uoc = 1
    tong_uoc = 0
    while uoc < b:
        if b % uoc == 0:
            tong_uoc += uoc
        uoc += 1
    return tong_uoc == b
    
def so_chinh_phuong(c):  
    so_chinh_phuong = c**0.5
    return so_chinh_phuong == int(so_chinh_phuong)

def so_Amstrong(d):
    so_luong_chu_so = len(str(d))
    d_nguyen = d    
    tong = 0

    while d > 0:
        tach_so = d % 10
        tong += tach_so ** so_luong_chu_so
        d = d // 10
        
    return tong == d_nguyen
    
# Nhập chuỗi và tách so sánh   
chuoi = input("Nhập vào chuoi:")
print("Chuỗi đã nhập là:", chuoi)
chuoi += " "
so_hien_tai = ""
ds_so = ""
s = 0
for ky_tu in chuoi:
    if ky_tu.isdigit():
        so_hien_tai += ky_tu   
    else:
        if so_hien_tai != "":
            n = int(so_hien_tai)
            if so_nguyen_to(n):
                s += n
                print(n,"Là số nguyên tố và S là:",s)
            elif so_hoan_hao(n):
                s -= n
                print(n,"Là số hoàn hảo và S là:",s )
            elif so_chinh_phuong(n):
                s /= n
                print(n,"Là số chính phương và S là:",s)
            elif so_Amstrong(n):
                s = 0
                print(n, "Là số Amstrong và S là:", s)
            else:
                print(n, "Không thuộc loại nào")

            ds_so += so_hien_tai + ","
            so_hien_tai = ""

if ds_so != "":
    ds_so = ds_so[:-1]
    print("Danh sách số đã tách là:", ds_so)
```
