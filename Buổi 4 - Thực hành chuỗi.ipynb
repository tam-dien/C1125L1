```python
### chuỗi nhập vào chỉ bao gồm ký tự chữ (không bao gồm ký tự số, ký tự đặc biệt,...)
### trong trường hợp chuỗi của mình là chuỗi rỗng ~~> False

chuoi = input("Nhập vào chuỗi: ")
ket_qua = chuoi != "" and not chuoi.islower()
print(ket_qua)
```

Nhập vào một chuỗi a và chuỗi b, tìm vị trí đầu tiên xuất hiện chuỗi b trong chuỗi a

```python
# chuoi_a = "xin chào"
# chuoi_b = "n"
# kết quả: 2

# chuoi_a = "xin chào các bạn, xin chào mọi người"
# chuoi_b = "chào"
# kết quả: 4

chuoi_a = input("Nhập vào chuỗi a:")
chuoi_b = input("Nhập vào chuỗi b:")

ket_qua = chuoi_a.find(chuoi_b)
print(chuoi_a)
print(chuoi_b)
print(ket_qua)
```

Nhập vào một chuỗi a, hãy xóa từ đầu tiên trong chuỗi a, rồi in ra màn hình

```python
### chuoi_a = "xin chào các bạn" ~~> "chào các bạn"

a = input("Nhập chuỗi a:")

b = a.find(" ")

ket_qua = a[b+1:]
print(a)
print(ket_qua)
```
